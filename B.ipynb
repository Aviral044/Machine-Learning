```python
A=np.array([[1,1],[1.5,4]])
B=np.array([2200,5050])
X=np.linalg.inv(A).dot(B)
print(X)
X=np.linalg.solve(A,B)
print(X)
#a+b=2200
#1.5a+4b=5050
```

```python
A_T=np.transpose(A)
print(A_T)
#Shortcut
print(A.T)

#DETERMINANT
np.linalg.det(A)
#Diagonal
np.diag(A)
```

```python
x=np.linspace(0,20,71500)
print(x)
A_signal=np.sin(x)
B_signal=np.cos(x)
plt.plot(x,A_signal,label='Sine Wave')
plt.plot(x,B_signal,label='Sine Wave')
plt.xlabel("Input")
plt.ylabel("Output")
plt.legend()
plt.show()
```

```python
#Donut DAta
#XOR SCATTERPLOT

X=np.random.rand(500,2)-0.5
Y=np.zeros(500)
for i in range(0,500):
    if(X[i][0]>0 and X[i][1]>0 or X[i][0]<0 and X[i][1]<0):
        Y[i]=1
    
        

plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
#Donut DAta
#XOR SCATTERPLOT
L=1000
R=np.random.rand(L)*2+2
T=np.linspace(0,np.pi*2,L)
X=R*np.sin(T)
Y=R*np.cos(T)

plt.scatter(X,Y)
plt.axis('square')
plt.show()
```

```python
L=1000
R=np.random.rand(L)*2+2
T=np.linspace(0,np.pi*2,L)
X=R*np.sin(T)
Y=R*np.cos(T)
Z=np.zeros(1000)
for i in range(0,1000):
    if(R[i]>3.0):
        Z[i]=1;

plt.scatter(X,Y,c=Z)
plt.axis('square')
plt.show()
```
